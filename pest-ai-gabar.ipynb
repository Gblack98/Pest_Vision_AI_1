```python
import os
import gc
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import EfficientNetV2B0
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.preprocessing import image_dataset_from_directory
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt

# Suppress excessive TF logging
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

# Enable mixed precision (bfloat16) for TPU/GPU
tf.keras.mixed_precision.set_global_policy('mixed_bfloat16')

# Initialize TPU strategy or fallback to GPU/CPU
try:
    # For TPU VM on Kaggle, specify 'local'
    resolver = tf.distribute.cluster_resolver.TPUClusterResolver(tpu='local')
    tf.config.experimental_connect_to_cluster(resolver)
    tf.tpu.experimental.initialize_tpu_system(resolver)
    strategy = tf.distribute.TPUStrategy(resolver)
    print("âœ… TPU detected and initialized")
except Exception as e:
    print(f"âš ï¸ TPU initialization failed: {e}")
    strategy = tf.distribute.MirroredStrategy()
    print("âœ… Using MirroredStrategy (GPU/CPU)")

print(f"ğŸš€ Number of replicas: {strategy.num_replicas_in_sync}")

# Global parameters
IMG_SIZE = (384, 384)
SEED = 42
EPOCH_HEAD = 5
EPOCH_FINE = 15
AUTOTUNE = tf.data.AUTOTUNE

# Directories
dataset_root = '/kaggle/input/niayes-crops-disease-v2'
train_dir = os.path.join(dataset_root, 'train')
val_dir   = os.path.join(dataset_root, 'val')
test_dir  = os.path.join(dataset_root, 'test')

CLASS_NAMES = sorted(os.listdir(train_dir))
NUM_CLASSES = len(CLASS_NAMES)
print(f"Categories detected: {CLASS_NAMES}")

# Compute batch size per replica
GLOBAL_BATCH_SIZE = 128 * strategy.num_replicas_in_sync

# Data pipeline

def build_dataset(directory, shuffle=False, augment=False):
    ds = image_dataset_from_directory(
        directory,
        image_size=IMG_SIZE,
        batch_size=GLOBAL_BATCH_SIZE,
        shuffle=shuffle,
        seed=SEED,
        label_mode='int',
        interpolation='bilinear'
    )

    def preprocess(image, label):
        image = tf.keras.applications.efficientnet_v2.preprocess_input(image)
        label = tf.one_hot(label, NUM_CLASSES)
        return image, label

    ds = ds.map(preprocess, num_parallel_calls=AUTOTUNE)

    if augment:
        augmentation = tf.keras.Sequential([
            layers.RandomFlip('horizontal'),
            layers.RandomRotation(0.15),
            layers.RandomZoom(0.1),
            layers.RandomContrast(0.1),
        ])
        ds = ds.map(lambda x, y: (augmentation(x, training=True), y), num_parallel_calls=AUTOTUNE)

    return ds.cache().prefetch(buffer_size=AUTOTUNE)

train_ds = build_dataset(train_dir, shuffle=True, augment=True)
val_ds   = build_dataset(val_dir)
test_ds  = build_dataset(test_dir)

# Calculate class weights
counts = np.array([len(os.listdir(os.path.join(train_dir, cls))) for cls in CLASS_NAMES])
labels = np.concatenate([np.full(c, i) for i, c in enumerate(counts)])
weights = compute_class_weight('balanced', classes=np.arange(NUM_CLASSES), y=labels)
class_weights = dict(enumerate(weights))
print("Class weights:", class_weights)

# Model construction under strategy scope
with strategy.scope():
    base_model = EfficientNetV2B0(
        include_top=False,
        weights='imagenet',
        input_shape=(*IMG_SIZE, 3),
        pooling='avg'
    )
    base_model.trainable = False

    inputs = layers.Input(shape=(*IMG_SIZE, 3))
    x = base_model(inputs, training=False)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.4)(x)
    outputs = layers.Dense(NUM_CLASSES, activation='softmax', dtype='float32')(x)
    model = models.Model(inputs, outputs)

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

model.summary()

# Callbacks
callbacks = [
    EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True),
    ModelCheckpoint('best_model.keras', monitor='val_accuracy', save_best_only=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=2, min_lr=1e-6)
]

# Phase 1: Train top layers
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCH_HEAD,
    class_weight=class_weights,
    callbacks=callbacks
)

# Phase 2: Fine-tuning
with strategy.scope():
    base_model.trainable = True
    for layer in base_model.layers[:100]:
        layer.trainable = False

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

history_fine = model.fit(
    train_ds,
    validation_data=val_ds,
    initial_epoch=history.epoch[-1],
    epochs=history.epoch[-1] + EPOCH_FINE,
    class_weight=class_weights,
    callbacks=callbacks
)

# Evaluation
loss, acc = model.evaluate(test_ds, verbose=0)
print(f"Final test accuracy: {acc*100:.2f}%")

y_true = np.concatenate([y for x, y in test_ds], axis=0)
y_pred = np.argmax(model.predict(test_ds), axis=1)

print(classification_report(y_true.argmax(axis=1), y_pred, target_names=CLASS_NAMES))

plt.figure(figsize=(12, 8))
cm = confusion_matrix(y_true.argmax(axis=1), y_pred)
plt.imshow(cm, interpolation='nearest', cmap='Blues')
plt.title('Confusion Matrix')
plt.colorbar()
plt.xticks(np.arange(NUM_CLASSES), CLASS_NAMES, rotation=90)
plt.yticks(np.arange(NUM_CLASSES), CLASS_NAMES)
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.tight_layout()
plt.show()

# Cleanup
tf.keras.backend.clear_session()
gc.collect()


# --- Save the final model ---

save_dir = '/kaggle/working/final_model'
os.makedirs(save_dir, exist_ok=True)



```

```python
# Sauvegarde au format Keras (.keras) et TensorFlow SavedModel
model.save(os.path.join(save_dir, 'model_final.keras'))
model.save(os.path.join(save_dir, 'model_final_savedmodel.keras'))

print(f"âœ… Model saved at: {save_dir}")

# --- Script to reload and test the model ---

# Chargement du modÃ¨le (Keras ou SavedModel)
model_reloaded = tf.keras.models.load_model(os.path.join(save_dir, 'model_final.keras'))
print("âœ… Model successfully reloaded!")

# Evaluation rapide sur le test set
loss_reload, acc_reload = model_reloaded.evaluate(test_ds, verbose=0)
print(f"Reloaded model test accuracy: {acc_reload*100:.2f}%")

# PrÃ©dictions reload pour confirmation
y_pred_reload = np.argmax(model_reloaded.predict(test_ds), axis=1)

print("\nReloaded model classification report:")
print(classification_report(y_true.argmax(axis=1), y_pred_reload, target_names=CLASS_NAMES))
```
